```python
import pandas as pd
from sklearn.model_selection import train_test_split
from zoofs import ParticleSwarmOptimization
import lightgbm as lgb
from sklearn.metrics import log_loss, accuracy_score, classification_report, f1_score
```

```python
df  = pd.read_csv('F:\Projects\Cap_github\Capstone\Data\set12.csv')
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df['label'] = df['label'].apply(lambda x: 0 if 'Brute' in str(x) else 1)
```

```python
X = df.drop(['label'],axis=1)
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.8, test_size = 0.2, shuffle=True,random_state = 0)
```

```python
def objective_function_topass(model,X_train, y_train, X_valid, y_valid):      
    model.fit(X_train,y_train)  
    P = log_loss(y_valid,model.predict_proba(X_valid))
    return P
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import f1_score

def objective_function_topass(model, X_train, y_train, X_valid, y_valid):      
    """
    Objective function for optimization. Trains the model and evaluates F1 score.
    Returns the negative F1 score to maximize it.
    """
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)  
    P = f1_score(y_valid, y_pred, average='weighted')  # Weighted F1 score
    return -P  # Negative F1 score to maximize it

# Assuming ParticleSwarmOptimization is implemented and imported
algo_object = ParticleSwarmOptimization(
    objective_function_topass,
    n_iteration=15,
    population_size=30,
    minimize=True
)

# Replace LightGBM model with XGBoost
xgb_model = XGBClassifier(
    eval_metric="logloss"  # Evaluation metric for XGBoost
)

# Fit the optimization algorithm
algo_object.fit(xgb_model, X_train, y_train, X_test, y_test, verbose=True)

```

```python
from zoofs import HarrisHawkOptimization

def objective_function_topass_harris(model, X_train, y_train, X_valid, y_valid):      
    model.fit(X_train, y_train)  
    y_pred = model.predict(X_valid)  # Use predicted class labels
    P = f1_score(y_valid, y_pred, average='weighted')  # Ensure the correct `average` is used
    return -P

# import an algorithm !  


# create object of algorithm
algo_object_haris=HarrisHawkOptimization(objective_function_topass_harris,
                                   n_iteration=10,
                                   population_size=30,
                                   minimize=True)

lgb_model = lgb.LGBMClassifier() 
# fit the algorithm
algo_object_haris.fit(lgb_model,X_train, y_train, X_test, y_test,verbose=True)

#plot your results
algo_object_haris.plot_history()

# extract the best  feature set
algo_object_haris.best_feature_list 
```

```python
from zoofs import DragonFlyOptimization
```

```python
def objective_function_topass_dragon(model, X_train, y_train, X_valid, y_valid):      
    model.fit(X_train, y_train)  
    y_pred = model.predict(X_valid)  # Use predicted class labels
    P = f1_score(y_valid, y_pred)  # Ensure the correct `average` is used
    return -P

algo_object_dragon=DragonFlyOptimization(objective_function_topass_dragon, n_iteration=15, population_size=40, method='linear',minimize=True)

lgb_model = lgb.LGBMClassifier() 
# fit the algorithm
algo_object_dragon.fit(lgb_model,X_train, y_train, X_test, y_test,verbose=True)
```

```python
from sklearn.feature_selection import SelectKBest, chi2

# Get the feature set from the optimization results
X_train_filtered = X_train[feat_drag]
X_test_filtered = X_test[feat_drag]

# Apply Chi-Square feature selection
chi2_selector = SelectKBest(chi2, k=15)  # Select top 20 features; adjust 'k' as needed
X_train_selected = chi2_selector.fit_transform(X_train_filtered, y_train)
X_test_selected = chi2_selector.transform(X_test_filtered)

# Get the selected feature indices
selected_indices = chi2_selector.get_support(indices=True)
selected_features = X_train.columns[selected_indices]
print("Selected Features after Chi-Square:", selected_features)

# Train the model on the reduced feature set
lgb_model = lgb.LGBMClassifier()
lgb_model.fit(X_train_selected, y_train)
y_pred = lgb_model.predict(X_test_selected)

# Evaluate the model
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
print("F1 Score with Chi-Square Selected Features:", f1)
```

```python
# Define the objective function
def objective_function_topass_dragon_chi(model, X_train, y_train, X_valid, y_valid):
    # Apply Chi-Square feature selection
    chi2_selector = SelectKBest(chi2, k=20)  # Select top 20 features, adjust 'k' as needed
    X_train_selected = chi2_selector.fit_transform(X_train, y_train)
    X_valid_selected = chi2_selector.transform(X_valid)
    
    # Fit the model
    model.fit(X_train_selected, y_train)
    y_pred = model.predict(X_valid_selected)
    
    # Compute F1 score
    P = f1_score(y_valid, y_pred)
    return -P  # Minimize the negative F1 score

# Initialize DragonFlyOptimization
algo_object_dragon_chi = DragonFlyOptimization(objective_function_topass_dragon,
                                           n_iteration=15,
                                           population_size=40,
                                           method='linear',
                                           minimize=True)

# Define the model
lgb_model = lgb.LGBMClassifier()

# Fit the DragonFly algorithm
algo_object_dragon_chi.fit(lgb_model, X_train, y_train, X_test, y_test, verbose=True)
```
